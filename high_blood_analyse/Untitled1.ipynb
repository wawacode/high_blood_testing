性别对高血压病的影响
第一个思路：男的有多少人，高血压病有多少，高血压病的人数/男的总人数==比率，比率就是男性的患病率
        女的有多少人，高血压病有多少，高血压病的人数/女的总人数==比率，比率就是 女性的患病率
        将两者的患病率用饼图表示。得出结论高血压病男性的患病比率大，还是女性的患病比率。.
第二个思路：高血压患者中去看，并不从整个样本中去看。高血压患者中男性的比率，满足2个条件，首先是高血压患者，其次是男性，这两个条件都满足，就是and操作，统计高血压患者中男性的比率的值。高血压患者女中女性的比率，满足2个条件，首先是高血压患者，其次是女性，这两个条件都满足，就是and操作，统计高血压患者中女性的比率的值。 
最后用饼图来分析结论。


```python
#第二个思路
#患病总人数求出。
ill_all=datas[datas["Hypertension"]==1.0]["Hypertension"].count()
```

```python
ill_all
```

```python
#患病条件datas["Hypertension"]==1.0
#男性条件datas["Sex"]==1.0
#两个条件都满足用& ,注意两边条件要用括号括起来
ill_man=datas[(datas["Hypertension"]==1.0)&(datas["Sex"]==1.0)]["Hypertension"].count()
```

```python
ill_man
```

```python
#患病条件datas["Hypertension"]==1.0
#男性条件datas["Sex"]==2.0
#两个条件都满足用& ,注意两边条件要用括号括起来
ill_woman=datas[(datas["Hypertension"]==1.0)&(datas["Sex"]==2.0)]["Hypertension"].count()
```

```python
ill_woman
```

```python
rate=round(ill_man/ill_all,2)
unrate=1-rate
```

```python
from pyecharts.charts import Pie
x=["患者中男性比率","患者中女性比率"]
y=[rate,1-rate]
pie=(
    Pie()
    .add("患者中男女之间的比率",[list(z) for z in zip(x,y)])
)
pie.render("患者中男女比例.html")
```

由上图，得知女性患病率比较高，男性患病率比较低，结论女性比男性更容易得高血压。

```python
#先求女性总人数
woman_all=datas[datas["Sex"]==2.0]["Hypertension"].count()
```

```python
#女性患病率
woman_ill=datas[(datas["Sex"]==2.0)&(datas["Hypertension"]==1.0)]["Hypertension"].count()
```

```python
woman_ill
```

```python
woman_rate=round(woman_ill/woman_all,2)
```

```python
woman_rate
```

```python
#画两个饼图,第一个男性里面患病率的饼图，第二个女性里面患病率的饼图
#先画男性
man_x=["男性患病率","男性未患病率"]
man_y=[man_rate,1-man_rate]
pie=(
    Pie()
    .add("男性高血压患病率",[list(z) for z in zip(man_x,man_y)])
)
pie.render("男性高血压患病率.html")
woman_x=["女性患病率","女性未患病率"]
woman_y=[woman_rate,1-woman_rate]
pie=(
    Pie()
    .add("女性高血压患病率",[list(z) for z in zip(woman_x,woman_y)])
)
pie.render("女性高血压患病率.html")
```

数据分析的结论：男性比女性患病率大。

#采用调查问卷，上网收集，线下收集,数据不典型。
数据不典型，机器学习：基本的训练集，测试集,交叉验证
逻辑回归+交叉验证，正确率是多少

```python
#交叉验证的结果
from sklearn.model_selection import cross_val_score
#分类用逻辑回归的方法
from sklearn.linear_model  import LogisticRegression
alg=LogisticRegression()
predictions=["Age","Sex","FHBP","Smoke","Drink","BMI"]
#计算用此数据进行训练和测试，用交叉验证方法的正确率。
score=cross_val_score(alg,datas[predictions],datas["Hypertension"],cv=3)
#输出交叉验证正确率的平均值  76.65%    如果正确率95%以上
score.mean()
```
